```python
fake = Faker()
num_users = 20000
file_name = "Brain_Rot_Analysis_Data.sql"
file_name = "Brain_Rot_New_Tables_Data.sql"
occupations = ['Student', 'Software Engineer', 'Content Creator', 'Doctor', 'Marketing Specialist', 'Unemployed', 'Call Center Agent']
platforms = ['TikTok', 'Instagram Reels', 'YouTube Shorts', 'Facebook', 'YouTube', 'Twitter']
```

```python
with open(file_name, 'w', encoding='utf-8') as f:
    f.write("USE BrainRotAnalysis;\nGO\n\n")
    for user_id in range(1, num_users + 1):
        age = random.randint(15, 45)
        gender = random.choice(['Male', 'Female'])
        occ = random.choice(occupations)
        sleep = round(random.uniform(4, 7), 1) if age < 25 else round(random.uniform(6, 9), 1)
        f.write(f"INSERT INTO Users (age, gender, occupation, avg_sleep_hours) VALUES ({age}, '{gender}', '{occ}', {sleep});\n")
        is_heavy = random.random() > 0.3
        if is_heavy:
            platform, content, usage, scroll = random.choice(['TikTok', 'Instagram Reels']), 'Short-form', random.randint(240, 600), random.randint(7, 10)
            attn, react, anxiety, boredom, mood = random.randint(5, 40), random.randint(500, 900), random.randint(7, 10), random.randint(1, 3), random.choice(['Low', 'Very Low'])
        else:
            platform, content, usage, scroll = random.choice(['YouTube', 'Facebook']), random.choice(['Long-form', 'Static-Posts']), random.randint(30, 150), random.randint(1, 5)
            attn, react, anxiety, boredom, mood = random.randint(60, 95), random.randint(200, 450), random.randint(1, 5), random.randint(6, 10), random.choice(['High', 'Neutral'])
        f.write(f"INSERT INTO Digital_Consumption (user_id, platform_name, content_format, daily_usage_minutes, scrolling_speed_index) VALUES ({user_id}, '{platform}', '{content}', {usage}, {scroll});\n")
        f.write(f"INSERT INTO Cognitive_Metrics (user_id, attention_span_score, memory_recall_accuracy, reaction_time_ms, test_date) VALUES ({user_id}, {attn}, {round(random.uniform(30, 99), 2)}, {react}, '{fake.date_between(start_date='-1y', end_date='today')}');\n")
        f.write(f"INSERT INTO Psychological_State (user_id, anxiety_level, loneliness_score, boredom_tolerance, mood_rating) VALUES ({user_id}, {anxiety}, {random.randint(1,10)}, {boredom}, '{mood}');\n")
```
